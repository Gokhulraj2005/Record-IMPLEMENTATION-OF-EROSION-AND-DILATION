```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_img():
    img = np.zeros((600, 600), dtype='uint8')   
    font = cv2.FONT_HERSHEY_SIMPLEX             
    cv2.putText(img, 'GOKHUL', (150, 300), font, 2, (255), 5, cv2.LINE_AA)
    return img
```

```python
def display_img(img):
    plt.title("image")
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
img = load_img()
display_img(img)
kernel = np.ones((6, 7), np.uint8)
```

```python
eroded_image = cv2.erode(img, kernel, iterations=1)
plt.imshow(cv2.cvtColor(eroded_image, cv2.COLOR_BGR2RGB)) 
plt.title("Eroded Image")
plt.axis('off')
```

```python
dilated_image = cv2.dilate(img, kernel, iterations=1)
plt.imshow(cv2.cvtColor(dilated_image, cv2.COLOR_BGR2RGB))
plt.title("Dilated Image")
plt.axis('off')
```
